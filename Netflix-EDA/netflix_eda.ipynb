# üé¨ Netflix Movie Analytics (1990s EDA)

This notebook explores Netflix movies released in the 1990s using exploratory data analysis (EDA).  
We analyze durations, genres, and identify patterns using Pandas and Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
netflix_df.info()
netflix_df.describe()
```

```python
netflix_movies = netflix_df[netflix_df["type"] == "Movie"]
netflix_movies.head()
```

```python
netflix_90s = netflix_movies[
    (netflix_movies["release_year"] >= 1990) &
    (netflix_movies["release_year"] <= 1999)
]

netflix_90s.head()
netflix_90s.shape
```

```python
duration = netflix_90s["duration"].mode()[0]
duration
```

```python
short_action = netflix_90s[
    (netflix_90s["genre"] == "Action") &
    (netflix_90s["duration"] < 90)
]

short_movie_count = short_action.shape[0]
short_movie_count
```

```python
netflix_movies = netflix_df[netflix_df["type"] == "Movie"]

netflix_90s = netflix_movies[
    (netflix_movies["release_year"] >= 1990) &
    (netflix_movies["release_year"] <= 1999)
]

netflix_90s.head(), netflix_90s.shape
```

```python
duration = netflix_90s["duration"].mode()[0]
duration
```

```python
short_action = netflix_90s[
    (netflix_90s["genre"] == "Action") &
    (netflix_90s["duration"] < 90)
]

short_movie_count = short_action.shape[0]
short_movie_count
```

```python
plt.figure(figsize=(10, 5))
plt.hist(netflix_90s["duration"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Movie Durations (1990s)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
netflix_90s["genre"].value_counts().plot(kind="bar", color="coral")
plt.title("Genre Distribution of 1990s Movies")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

## üìå Conclusion

- The most common movie duration in the 1990s was **`{duration}` minutes**.
- There were **`{short_movie_count}`** short action movies (< 90 min).
- Most movie durations fall between 80‚Äì120 minutes.
- Drama, Action, and Comedy appear to be the most represented genres.
